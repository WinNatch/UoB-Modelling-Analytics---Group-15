**DEA - Main Dataset**

```python
!pip install pulp
import pulp
!apt-get install -y -qq glpk-utils
from pulp import GLPK
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/WinNatch/UoB-Modelling-Analytics--Group-15/main/DEA_Main.csv'
data = pd.read_csv(url)

weights_matrix = np.empty((14,9))
for k in range(9):
    dea_model = pulp.LpProblem("F1_Analysis", pulp.LpMaximize)
    m1 = pulp.LpVariable("u1", lowBound=0, upBound=None, cat='Continuous')
    m2 = pulp.LpVariable("u2", lowBound=0, upBound=None, cat='Continuous')
    m3 = pulp.LpVariable("u3", lowBound=0, upBound=None, cat='Continuous')
    m4 = pulp.LpVariable("u4", lowBound=0, upBound=None, cat='Continuous')
    m5 = pulp.LpVariable("u5", lowBound=0, upBound=None, cat='Continuous')
    m6 = pulp.LpVariable("u6", lowBound=0, upBound=None, cat='Continuous')
    n1 = pulp.LpVariable("v1", lowBound=0, upBound=None, cat='Continuous')
    n2 = pulp.LpVariable("v2", lowBound=0, upBound=None, cat='Continuous')
    n3 = pulp.LpVariable("v3", lowBound=0, upBound=None, cat='Continuous')
    n4 = pulp.LpVariable("v4", lowBound=0, upBound=None, cat='Continuous')
    n5 = pulp.LpVariable("v5", lowBound=0, upBound=None, cat='Continuous')
    n6 = pulp.LpVariable("v6", lowBound=0, upBound=None, cat='Continuous')
    n7 = pulp.LpVariable("v7", lowBound=0, upBound=None, cat='Continuous')
    n8 = pulp.LpVariable("v8", lowBound=0, upBound=None, cat='Continuous')

    # Objective Function
    dea_model += m1*data["Y1"][k] + m2*data["Y2"][k] + m3*data["Y3"][k] + m4*data["Y4"][k] + m5*data["Y5"][k] + m6*data["Y6"][k]

    # Constraint
    dea_model += n1*data["X1"][k] + n2*data["X2"][k] + n3*data["X3"][k] + n4*data["X4"][k] + n5*data["X5"][k] + n6*data["X6"][k] + n7*data["X7"][k] + n8*data["X8"][k] == 1

    for i in range(9):
        dea_model += m1*data["Y1"][i] + m2*data["Y2"][i] + m3*data["Y3"][i] + m4*data["Y4"][i] + m5*data["Y5"][i] + m6*data["Y6"][i] - n1*data["X1"][i] - n2*data["X2"][i] - n3*data["X3"][i] - n4*data["X4"][i] - n5*data["X5"][i] - n6*data["X6"][i] - n7*data["X7"][i] - n8*data["X8"][i] <= 0

    # Solve The Model
    status = dea_model.solve()
    print("Model Status:{}".format(pulp.LpStatus[dea_model.status]))

    # Save Variables in Matrix
    list_weights=[]
    for v in dea_model.variables():
	      print(v.name, "=", v.varValue)
    for v in dea_model.variables():
        list_weights = list_weights + [v.varValue]
    for n in range(14):
        value = list_weights[n]
        weights_matrix[n,k] = value

    print("Objective=", pulp.value(dea_model.objective))

    print("\nConstraint")
    for name, c in dea_model.constraints.items():
	      print(name, ":", c)
```

**Cross Efficiency & Arithmetic Cross-efficiency**

```python
from statistics import mean
import pandas as pd

# Create Cross Efficiency Table and Calculate Cross-efficiency Score
cross_efficiency_table = np.empty((9,9))
avg_efficiency = []
for row in range(9):
    row_values = []
    for column in range(9):
        value = ((weights_matrix[0,column]*data["Y1"][row])+(weights_matrix[1,column]*data["Y2"][row])+(weights_matrix[2,column]*data["Y3"][row])+(weights_matrix[3,column]*data["Y4"][row])+(weights_matrix[4,column]*data["Y5"][row])+(weights_matrix[5,column]*data["Y6"][row]))/((weights_matrix[6,column]*data["X1"][row])+(weights_matrix[7,column]*data["X2"][row])+(weights_matrix[8,column]*data["X3"][row])+(weights_matrix[9,column]*data["X4"][row])+(weights_matrix[10,column]*data["X5"][row])+(weights_matrix[11,column]*data["X6"][row])+(weights_matrix[12,column]*data["X7"][row])+(weights_matrix[13,column]*data["X8"][row]))
        cross_efficiency_table[row,column] = value
        row_values += [value]

    # Calculate Arithmetic mean of Cross-efficiency score
    average = mean(row_values)
    avg_efficiency += [average]

average_list_rounded = [ '%.3f' % elem for elem in avg_efficiency ]

F1_Teams = data['F1-Team'].tolist()

for team in F1_Teams:
    index = F1_Teams.index(team)
    print("Arithmetic Cross-efficiency score for",team,"=",average_list_rounded[index])

F1_cross_efficiency = cross_efficiency_table.round(decimals=2)

F1_cross_efficiency_table = pd.DataFrame(F1_cross_efficiency)
F1_cross_efficiency_table.insert(0, 'F1-Teams', F1_Teams)
set_column_names = ['F1-Teams'] + [team + " weight" for team in F1_Teams]
F1_cross_efficiency_table.columns = set_column_names

display(F1_cross_efficiency_table)

F1_cross_efficiency_table.to_csv("Cross_Eff_DEA_Result.csv", index=False)
```

**GEOMETRIC MEAN CROSS-EFFICIENCY**

```python
import math
import numpy as np

# Convert to a numpy array for convenience
matrix = np.array(F1_cross_efficiency)

# Fixing the zero efficiency score
matrix[4,1] = 1e-6
matrix[5,1] = 1e-6
matrix[6,1] = 1e-6
matrix[7,1] = 1e-6
matrix[8,1] = 1e-6
matrix[7,4] = 1e-6

# Compute geometric mean for each DMU (rows)
geometric_means = [math.prod(matrix[i, :]) ** (1 / matrix.shape[1]) for i in range(matrix.shape[0])]

# Display results
for dmu, gm in enumerate(geometric_means, start=0):
    team_name = F1_Teams[dmu]
    print(f"Geometric Mean efficiency for {team_name}: {round(gm, 3):.3f}")
```

**CONSENSUS CROSS-EFFICIENCY**

```python
F1_cross_efficiency
```

```python
# Set a DMU as CRP (Comparative Reference Point)
CRP_DMU = F1_cross_efficiency[1]

# Rescaling by divide each column by the corresponding value in CRP row
Rescale_Matrix = np.zeros_like(F1_cross_efficiency)
for i in range(F1_cross_efficiency.shape[1]):
    Rescale_Matrix[:, i] = F1_cross_efficiency[:, i] / CRP_DMU[i]

Rescale_Matrix = np.round(Rescale_Matrix, 2)

Rescale_Matrix
```

```python
import math

# Fixing the value of zero
Rescale_Matrix[4,1] = 1e-6
Rescale_Matrix[5,1] = 1e-6
Rescale_Matrix[6,1] = 1e-6
Rescale_Matrix[7,1] = 1e-6
Rescale_Matrix[8,1] = 1e-6
Rescale_Matrix[7,4] = 1e-6

# Compute geometric mean for each DMU (rows)
geometric_mean = [math.prod(Rescale_Matrix[i, :]) ** (1 / Rescale_Matrix.shape[1]) for i in range(Rescale_Matrix.shape[0])]

# Initialize diviation as a list to store the values
root_sum_squared_list = []

# Iterate over each column
for col in range(Rescale_Matrix.shape[1]):
  column = Rescale_Matrix[:, col]  # Select the current column
  mean = geometric_mean[col]  # Access corresponding geometric mean

  # Calculate square root of squared deviations and sum
  deviations = math.sqrt(sum((x - mean)**2 for x in column))

  # Append the root of sum of squared deviations for the column
  root_sum_squared_list.append(deviations)

# Convert the string values to float
root_sum_squared_list = [float(x) for x in root_sum_squared_list]

root_sum_squared_list
```

```python
# Calculate ECD (Evaluation Consensus Degree) for each value
ECD = [1 / (1 + x) for x in root_sum_squared_list]

ECD
```

```python
import numpy as np

# Calculate the sum of ECD
total_power = sum(ECD)

# Calculate the CEC score for each DMU
CEC_Results = []
for i in range(len(ECD)):
  row = Rescale_Matrix[i]
  product = np.prod(np.power(row, ECD[i]))
  result = np.power(product, 1/total_power)
  CEC_Results.append(result)

CEC_Results = np.array(CEC_Results)

for team in F1_Teams:
    index = F1_Teams.index(team)
    print("CEC score for",team,"=",CEC_Results[index])
```

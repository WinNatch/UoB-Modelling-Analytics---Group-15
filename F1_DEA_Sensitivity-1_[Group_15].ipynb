**Sensitivity Test 1 - SPENDING EFFICIENCY**

```python
!pip install pulp
import pulp
!apt-get install -y -qq glpk-utils
from pulp import GLPK
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/WinNatch/UoB-Modelling-Analytics--Group-15/main/DEA_SA_1.csv'
data_sa_1 = pd.read_csv(url)

weights_matrix_1 = np.empty((9,9))
for k in range(9):
    sa1_dea_model = pulp.LpProblem("F1_Analysis", pulp.LpMaximize)
    m1 = pulp.LpVariable("m1", lowBound=0, upBound=None, cat='Continuous')
    m2 = pulp.LpVariable("m2", lowBound=0, upBound=None, cat='Continuous')
    m3 = pulp.LpVariable("m3", lowBound=0, upBound=None, cat='Continuous')
    m4 = pulp.LpVariable("m4", lowBound=0, upBound=None, cat='Continuous')
    m5 = pulp.LpVariable("m5", lowBound=0, upBound=None, cat='Continuous')
    m6 = pulp.LpVariable("m6", lowBound=0, upBound=None, cat='Continuous')
    n1 = pulp.LpVariable("n1", lowBound=0, upBound=None, cat='Continuous')
    n2 = pulp.LpVariable("n2", lowBound=0, upBound=None, cat='Continuous')
    n3 = pulp.LpVariable("n3", lowBound=0, upBound=None, cat='Continuous')

    sa1_dea_model += m1*data_sa_1["Y1"][k] + m2*data_sa_1["Y2"][k] + m3*data_sa_1["Y3"][k] + m4*data_sa_1["Y4"][k] + m5*data_sa_1["Y5"][k] + m6*data_sa_1["Y6"][k]

    sa1_dea_model += n1*data_sa_1["X1"][k] + n2*data_sa_1["X2"][k] + n3*data_sa_1["X3"][k] == 1

    for i in range(9):
        sa1_dea_model += m1*data_sa_1["Y1"][i] + m2*data_sa_1["Y2"][i] + m3*data_sa_1["Y3"][i] + m4*data_sa_1["Y4"][i] + m5*data_sa_1["Y5"][i] + m6*data_sa_1["Y6"][i] - n1*data_sa_1["X1"][i] - n2*data_sa_1["X2"][i] - n3*data_sa_1["X3"][i] <= 0

    status = sa1_dea_model.solve()
    print("Model Status:{}".format(pulp.LpStatus[sa1_dea_model.status]))

    list_weights=[]
    for v in sa1_dea_model.variables():
	      print(v.name, "=", v.varValue)
    for v in sa1_dea_model.variables():
        list_weights = list_weights + [v.varValue]
    for n in range(9):
        value = list_weights[n]
        weights_matrix_1[n,k] = value

    print("Objective=", pulp.value(sa1_dea_model.objective))

    print("\nConstraint")
    for name, c in sa1_dea_model.constraints.items():
	      print(name, ":", c)
```

```python
from statistics import mean
import pandas as pd

sa1_cross_efficiency_table = np.empty((9,9))
sa1_avg_efficiency = []
for row in range(9):
    row_values = []
    for column in range(9):
        value = ((weights_matrix_1[0,column]*data_sa_1["Y1"][row])+(weights_matrix_1[1,column]*data_sa_1["Y2"][row])+(weights_matrix_1[2,column]*data_sa_1["Y3"][row])+(weights_matrix_1[3,column]*data_sa_1["Y4"][row])+(weights_matrix_1[4,column]*data_sa_1["Y5"][row])+(weights_matrix_1[5,column]*data_sa_1["Y6"][row]))/((weights_matrix_1[6,column]*data_sa_1["X1"][row])+(weights_matrix_1[7,column]*data_sa_1["X2"][row])+(weights_matrix_1[8,column]*data_sa_1["X3"][row]))
        sa1_cross_efficiency_table[row,column] = value
        row_values += [value]
    average = mean(row_values)
    sa1_avg_efficiency += [average]

sa1_average_list_rounded = [ '%.2f' % elem for elem in sa1_avg_efficiency ]

F1_Teams = data_sa_1['F1-Team'].tolist()

for team in F1_Teams:
    index = F1_Teams.index(team)
    print("Arithmetic Cross-efficiency score for",team,"=",sa1_average_list_rounded[index])

SA1_cross_efficiency_table = sa1_cross_efficiency_table.round(decimals=2)

F1_SA_1 = pd.DataFrame(SA1_cross_efficiency_table)
F1_SA_1.insert(0, 'F1-Teams', F1_Teams)
set_column_names = ['F1-Teams'] + [team + " weight" for team in F1_Teams]
F1_SA_1.columns = set_column_names

display(F1_SA_1)

F1_SA_1.to_csv("SA_1_Result.csv", index=False)
```

**GEOMETRIC MEAN CROSS-EFFICIENCY**

```python
import math
import numpy as np

# Convert to a numpy array for convenience
matrix = np.array(SA1_cross_efficiency_table)

# Fixing the zero efficiency score
matrix[2,1] = 1e-6
matrix[5,1] = 1e-6
matrix[6,1] = 1e-6
matrix[7,1] = 1e-6
matrix[7,4] = 1e-6

# Compute geometric mean for each DMU (rows)
geometric_means = [math.prod(matrix[i, :]) ** (1 / matrix.shape[1]) for i in range(matrix.shape[0])]

# Display results
for dmu, gm in enumerate(geometric_means, start=0):
    team_name = F1_Teams[dmu]
    print(f"Geometric Mean efficiency for {team_name}: {round(gm, 3):.3f}")
```

**CONSENSUS CROSS-EFFICIENCY**

```python
import numpy as np
import math

# Get the CRP
second_row = SA1_cross_efficiency_table[8]

# Create a new array and divide each column by the corresponding value in the second row
Rescale_Matrix = np.zeros_like(SA1_cross_efficiency_table)
for i in range(SA1_cross_efficiency_table.shape[1]):
    Rescale_Matrix[:, i] = SA1_cross_efficiency_table[:, i] / second_row[i]

Rescale_Matrix = np.round(Rescale_Matrix, 2)

# Fixing the zero efficiency score
Rescale_Matrix[2,1] = 1e-6
Rescale_Matrix[5,1] = 1e-6
Rescale_Matrix[6,1] = 1e-6
Rescale_Matrix[7,1] = 1e-6
Rescale_Matrix[7,4] = 1e-6

# Compute geometric mean for each DMU (rows)
geometric_mean = [math.prod(Rescale_Matrix[i, :]) ** (1 / Rescale_Matrix.shape[1]) for i in range(Rescale_Matrix.shape[0])]

# Initialize sum_squared as a list to store the values
root_sum_squared_list = []

# Iterate over each column
for col in range(Rescale_Matrix.shape[1]):
  column = Rescale_Matrix[:, col]  # Select the current column
  mean = geometric_mean[col]  # Access corresponding geometric mean

  # Calculate square root of squared deviations and sum
  deviations = math.sqrt(sum((x - mean)**2 for x in column))

  # Append the root of sum of squared deviations for the column
  root_sum_squared_list.append(deviations)

# Convert the string values to float
root_sum_squared_list = [float(x) for x in root_sum_squared_list]

# Calculate 1/(1+x) for each value
ECD = [1 / (1 + x) for x in root_sum_squared_list]

# Calculate the sum of all powers
total_power = sum(ECD)

# Calculate the product of each row, raised to the power of the total power
CEC_Results = []
for i in range(len(ECD)):
  row = Rescale_Matrix[i]
  product = np.prod(np.power(row, ECD[i]))
  result = np.power(product, 1/total_power)
  CEC_Results.append(result)

CEC_Results = np.array(CEC_Results)  # Convert to NumPy array

for team in F1_Teams:
    index = F1_Teams.index(team)
    print("CEC score for",team,"=",CEC_Results[index])
```
